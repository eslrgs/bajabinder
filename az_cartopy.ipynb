```python
import xarray as xr
import cartopy.crs as ccrs
import cartopy
import matplotlib.pyplot as plt
%matplotlib inline
ds = xr.open_dataset('/Users/euan-soutter/.gmt/server/earth_relief_60m.grd')
ds
z = ds.z.sel()
fig = plt.figure(figsize=(9,6))
ax = plt.axes(projection=ccrs.Robinson())
ax.coastlines()
ax.gridlines()
z.plot(ax=ax, cmap='terrain',transform=ccrs.PlateCarree(), cbar_kwargs={'boundaries':
                                                                        [-5000,0,5000],'shrink': 0.4})
```
