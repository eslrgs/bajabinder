```python
import pygmt as pg
import os
import pandas as pd
import numpy as np
```

```python
os.chdir("/Users/euan-soutter/Desktop/Python/Field_Data/Azerbaijan/Data/") # Sets working directory
df = pd.read_csv("az_loc.csv")
print(df) # Displays statistical descriptions of df
```

```python
region = [48.12, 48.43, 41.17, 41.23]
fig = pg.Figure()
fig.coast(region = region, projection='M20c', frame='a', land='grey')
fig.plot(x = df.long, y=df.lat, color='white',style='c0.2c')
fig.show()
```

```python
topo = pg.datasets.load_earth_relief('02m')
region=[35,55,37,46]
fig = pg.Figure()
fig.basemap(R = region, projection='N20c', frame='a')
fig.grdimage(topo, cmap='geo')
fig.colorbar(position='JCR+v', frame=['x2000','y+lm'])
fig.show()
```
